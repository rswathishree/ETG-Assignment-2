# Summer Olympics Data Analysis Assignment

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("summer.csv")
```

```python
#function for writing values of each bar in bar graph
def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i, y[i], y[i], ha = 'center')
```

```python
df.head()
```

### 1. In how many cities Summer Olympics is held so far?

```python
a=len(df['City'].unique()) #displays the number of cities in which Olympics held
print("Summer Olympics is held at ",a," different Cities")
```

### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
data = []
df1=df[df['Medal']=='Gold']
for sp in df1['Sport'].unique():
    data.append([sp , len(df1[df1['Sport']  ==sp])])
a=pd.DataFrame(data,columns = ['Sport','number of Gold Medals']).sort_values(by='number of Gold Medals', ascending=False).head()
a.plot(x = 'Sport', y = 'number of Gold Medals', kind = 'bar', figsize = (5,5),color='Pink')
x=a['Sport']
y=[]
for no in a['number of Gold Medals']:
    y.append(no)
addlabels(x, y)
```

### 3. Which sport is having most number of medals so far? (Top 5)

```python
data = []
for sp in df['Sport'].unique():
    data.append([sp , len(df[df['Sport']  ==sp])])
a=pd.DataFrame(data,columns = ['Sport','number of medals']).sort_values(by='number of medals', ascending=False).head()
a.plot(x = 'Sport', y = 'number of medals', kind = 'bar', figsize = (5,5),color="green")
x=a['Sport']
y=[]
for no in a['number of medals']:
    y.append(no)
addlabels(x, y)
```

### 4. Which player has won most number of medals? (Top 5)

```python
data = []
for name in df['Athlete'].unique():
    data.append([name , len(df[df['Athlete']  ==name])])
a=pd.DataFrame(data,columns = ['Athlete','No of Medals']).sort_values(by='No of Medals', ascending=False).head()
a.plot(x = 'Athlete', y = 'No of Medals', kind = 'bar', figsize = (5,5),color='Violet')
x=a['Athlete']
y=[]
for no in a['No of Medals']:
    y.append(no)
addlabels(x, y)
```

### 5. Which player has won most number Gold Medals of medals? (Top 5)

```python
data = []
df1=df[df['Medal']=='Gold']
for name in df1['Athlete'].unique():
    data.append([name , len(df1[df1['Athlete']  ==name])])
a=pd.DataFrame(data,columns = ['Athlete','No of Gold Medals']).sort_values(by='No of Gold Medals', ascending=False).head()
a.plot(x = 'Athlete', y = 'No of Gold Medals', kind = 'bar', figsize = (5,5),color="darkBlue")
x=a['Athlete']
y=[]
for no in a['No of Gold Medals']:
    y.append(no)
addlabels(x, y)
```

### 6. In which year India won first Gold Medal in Summer Olympics?

```python
df1=df[df['Medal']=='Gold']
df1=df1[df1['Country']=='IND']
print("India has won the first Gold medal in",str(df1['Year'].head(1)).split("    ")[1].split("\n")[0])
```

### 7. Which event is most popular in terms on number of players? (Top 5)

```python
data = []
for ev in df['Event'].unique():
    data.append([ev , len(df[df['Event']  ==ev])])
a=pd.DataFrame(data,columns = ['Event','Count of players']).sort_values(by='Count of players', ascending=False).head()
a.plot(x = 'Event', y = 'Count of players', kind = 'bar', figsize = (5,5), color="maroon")
x=a['Event']
y=[]
for no in a['Count of players']:
    y.append(no)
addlabels(x, y)
```

### 8. Which sport is having most female Gold Medalists? (Top 5)

```python
data = []
df1=df[df['Gender']=='Women']
df1=df1[df1['Medal']=='Gold']
for name in df1['Sport'].unique():
    data.append([name , len(df1[df1['Sport']  ==name])])
a=pd.DataFrame(data,columns = ['Sport','No of Gold Medals- Women']).sort_values(by='No of Gold Medals- Women', ascending=False).head()
a.plot(x = 'Sport', y = 'No of Gold Medals- Women', kind = 'bar', figsize = (5,5),color="orange")
x=a['Sport']
y=[]
for no in a['No of Gold Medals- Women']:
    y.append(no)
addlabels(x, y)
```
